```python
import numpy as np
import matplotlib.pyplot as mtp 
import pandas as pd 
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv("datasets/OnlineRetail.csv",encoding ='latin1')
dataset.head()
```

```python
X= dataset.iloc[:,[3,5]].values
```

```python
wcss_list = []

for i in range (1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state = 42)
    kmeans.fit(X)
    wcss_list.append(kmeans.inertia_)
    
```

```python
mtp.plot(range(1,11), wcss_list)
mtp.title("The Elbow Method Graph")
mtp.xlabel("Number of Clusters")
mtp.ylabel("wcss_list")
mtp.show()
```

```python
kmeans = KMeans(n_clusters = 3,init='k-means++',random_state = 42)
y_predict = kmeans.fit_predict(X)
```

```python
mtp.scatter(X[y_predict == 0,0],X[y_predict == 0,1], s = 100,c = 'red',label='Cluster 01')
mtp.scatter(X[y_predict == 1,0],X[y_predict == 1,1], s = 100,c = 'green',label='Cluster 02')
mtp.scatter(X[y_predict == 2,0],X[y_predict == 2,1], s = 100,c = 'blue',label='Cluster 03')

mtp.title("Cluster of Online Retail")
mtp.xlabel("Quantity")
mtp.ylabel("Unit Price")
mtp.legend()
mtp.show()
```
