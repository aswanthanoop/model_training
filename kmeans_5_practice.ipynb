```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as mtp 
```

```python
data = pd.read_csv("datasets/Mall_Customers.csv")
data.head()
```

```python
X = data.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss_list = []

for i in range (1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss_list.append(kmeans.inertia_)
    
```

```python
mtp.plot(range(1,11),wcss_list)
mtp.title("The Elbow Graph")
mtp.xlabel("NUmber of Clusters")
mtp.ylabel("wcss_list")
mtp.show()
```

```python
kmeans = KMeans(n_clusters = 5,init = 'k-means++', random_state = 42)
y_pred = kmeans.fit_predict(X)
```

```python
mtp.scatter(X[y_pred ==0,0], X[y_pred == 0,1], s = 100, c = 'red', label = "Cluster 01")
mtp.scatter(X[y_pred == 1,0],X[y_pred == 1,1], s = 100 , c = 'green', label = "Cluster 02")
mtp.scatter(X[y_pred == 2,0],X[y_pred == 2,1], s = 100 , c = 'yellow', label = "Cluster 03")
mtp.scatter(X[y_pred == 3,0],X[y_pred == 3,1], s = 100 , c = 'blue', label = "Cluster 04")
mtp.scatter(X[y_pred == 4,0],X[y_pred == 4,1], s = 100 , c = 'orange', label = "Cluster 05")
mtp.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s = 100 ,c = "black", label = "Centroids")

mtp.title("Cluster of Customers")
mtp.xlabel("Annual Income (k$)")
mtp.ylabel("Spending score (1 - 100)")
mtp.legend()
mtp.show()
```
